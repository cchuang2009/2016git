Benefit from the javascript library, <code style="color:lightblue">d3.js</code>, experimental mpld3 avails the visualization functionalities:

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
````

````python
fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))
ax.grid(color='white', linestyle='solid')

N = 50
ax.scatter(np.random.normal(size=N),
                     np.random.normal(size=N),
                     c=np.random.random(size=N),
                     s = 1000 * np.random.random(size=N),
                     alpha=0.3,
                     cmap=plt.cm.jet)

ax.set_title("Matplotlib Scatter Plot", size=18);
````

````python
from mpld3 import plugins

fig, ax = plt.subplots(subplot_kw=dict(axisbg='#EEEEEE'))
ax.grid(color='white', linestyle='solid')

N = 50
scatter = ax.scatter(np.random.normal(size=N),
                     np.random.normal(size=N),
                     c=np.random.random(size=N),
                     s = 1000 * np.random.random(size=N),
                     alpha=0.3,
                     cmap=plt.cm.jet)

ax.set_title("D3 Scatter Plot (with tooltips!)", size=20)

labels = ['point {0}'.format(i + 1) for i in range(N)]
fig.plugins = [plugins.PointLabelTooltip(scatter, labels)]
````

Interact
---
programmable, visualisation, and interactivive widgets forms powerful UI environment.  <code style="color:lightblue">ipywidgets</code> suits our needs!

````python
from ipywidgets import widgets

import mpld3         # Notice that as you hover over the plot, a toolbar

def on_value_change(value):
    fig, ax = plt.subplots()
    x, y = np.random.normal(size=(2, 100))
    color, size = np.random.random((2, 100))

    ax.scatter(x[:value], y[:value], c=color[:value], s=500 * size, alpha=0.3)
    ax.axis((-3,3,-3,3))
    ax.grid(color='lightgray', alpha=0.7)


widgets.interact(on_value_change, value = widgets.IntSlider(min=1,value=20))
````
