Suppose that a sequence, $\left\{ a_n \right\}^{\infty}_{n = 0}$,  satisfies the following recursive formula:
$$ a_{n + 1} = f \left( a_n \right) $$
for some function $f \left( x \right)$ and $n = 1, 2, 3, \cdots$. 

Every element in the sequence can be represented in the form of composed function $f^n$ as following:

$$ \begin{eqnarray*} a_n & = & f \left( a_{n - 1} \right)\\ & = & f \left( f \left( a_{n - 2} \right) \right) = f^2 \left( a_{n - 2} \right)\\ & = & \cdots\\ & = & f^n \left( a_0 \right) \end{eqnarray*} $$

Periodic Point
---
Suppose that the sequence, $\left\{ a_n \right\}^{\infty}_{n = 0}$, satisfies:
 
$$a_{n + 1} = f \left( a_n \right)$$

$a_0$ is called **periodic point** with cycle of period
  ${\color{red} m }$ if there exists $m \in \mathbb{N}$ such that
  $$ {\color{red}{a_m}}=f \left( a_{m-1} \right) =f^2(a_{m-2})=\cdots=f^{m-1}(a_1)=f^m(a_0)= {\color{red}{a_0}}. $$
  
And $a_0$ is called **fixed point** if $f \left( a_0 \right) = a_0$.



<h2>Example:</h2>

Suppose that $f \left( x \right) = 2 / x$, $x \neq 0$.

Define a sequence, $\left\{ a_n \right\}^{\infty}_{n = 0}$ as  follows:
$$
a_n = \left\{ \begin{array}{ll}
         a & \text{if } n = 0,\\
         f \left( a_{n - 1} \right) & \text{if } n \neq 0.
       \end{array} \right. 
$$       


Since $f \left( a \right) = 2 / a$ and $f \left( 2 / a \right) = a$, the sequence could be also listed as follows:     
$$    
    \left\{ {\color{red} a}, 2 / a, {\color{red} a}, 2 / a, \cdots \right\}
$$
    Note that $f^{{\color{red} 2}} \left( x \right) = f \left( f \left(x\right) \right) = x$; Then there is one point with period 2 if $a\ne0$.

Suppose that the sequence is convergent, its limit can be found out by followings:
$$
\begin{eqnarray*}
  L & = & \lim_{n \rightarrow \infty} f^n \left( x \right)\\
  & = & \lim_{n \rightarrow \infty} f^{n + 1} \left( x \right) = f \left(
  \lim_{n \rightarrow \infty} f^n \left( x \right) \right)\\
  & = & f \left( L \right) = 2 / L\\
  & \Rightarrow & L^2 = 2\\
  & \Rightarrow & L = \pm \sqrt{2}
\end{eqnarray*}
$$
i.e. limit is one of $\pm \sqrt{2}$, depending on the sign of initial value, $a$.


Now consider the image:
$$
    f:\mathbb{R} \to \mathbb{R}
$$ 
where $\mathbb{R}$ is the set of all the real numbers.
<p>

- $a=0$: the limit of $a_n$ fails to exist since $2/0$ fails to exist.</div>
- $a\neq0 \text{ or }\pm\sqrt2$ (Case of points with period 2): 
$$   
\begin{eqnarray*}
        a_0 &=& a, \\
        a_1 &=& f(a_0)=f(a)=2/a\neq a, \\   
        a_2 &=& f(a_1)=f(2/a)=a,  \\
            &\cdots& \\
\end{eqnarray*}    
$$
- $a=\pm\sqrt2$ (Fixed Points Case): $a_n=\pm\sqrt2$ 
 for $n=0,1,2,3,\cdots$


**Conclusion:**
Only at $a=\pm\sqrt2$, the sequence is convergent.

## Fibonicci Sequence: 

The terms of the sequences are in the form as follows:
$$ f_0=f_1=1,f_2=f_0+f_1=2,⋯,f_{n+1}=f_n+f_{n−1},n=1,2,3,\cdots$$
This means:
$${f_n}=1,1,2,3,5,8,13,21,34,\cdots$$

## What is $f_n$?

Theoretically, 
$$ f_n= A_1 \left(\frac{1+\sqrt5}{2}\right)^n+A_2\left(\frac{1-\sqrt5}{2}\right)^n$$
since $f_n= A r^n$ is the fundamental solution of the Fibonicci equation:
$$ 
\begin{equation}
f_{n+1}=f_n+f_{n-1}\\
ar^{n+1}=ar^n+ar^{r-1}\\
r^2=r+1\\
r=\frac{1+\sqrt5}{2},\frac{1-\sqrt5}{2}
\end{equation}
$$

Consider $\mathbf{f_\mu(x) = \mu x (1-x)}$ for $\mathbf{x}\in[0,1]$.

**Fact**
$$\mathbf{f_4(x)=4x(1-x):[0,1]\to[0,1]}$$

````python
plt.figure(figsize=(6,4))
x=np.linspace(0,1,101)
plt.plot(x,4*x*(1-x))
````

````python
# define function
def f(x, mu):
    return mu*x*(1.-x)
````

````python
def Iterate(g, x0, N, mu):
    for i in range(N):
        x0 = g(x0, mu)
    return x0
````

````python
def IterateList(g, x0, N, mu):
    xs = [x0]
    for i in range(N-1):
        x0 = g(x0, mu)
        xs.append(x0)
    return xs
````

````python
def Bifurcation(g, x0, nTransient, nCycle, muArray, showPlot=True):
    muMatrix = []
    xMatrix = []
    for mu in muArray:
        muMatrix.append([mu]*nCycle)
        xMatrix.append(IterateList(g, Iterate(g, x0, nTransient, mu),nCycle, mu))
    return muMatrix, xMatrix
````
